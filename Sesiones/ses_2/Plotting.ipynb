¿Fácil, no? Vamos a hacer alguna cosa interesante que modifique el funcionamiento por defecto de la función.  

Por ejemplo, añadimos colores

```r
plot(test, col="red")
```

¿Cuántos colores podemos utilizar? 

¿Hay que saberse los nombres de todos los colores en inglés?

```r
showCols <- function(cl=colors(), bg = "grey",
                     cex = 0.75, rot = 30) {
    m <- ceiling(sqrt(n <-length(cl)))
    length(cl) <- m*m; cm <- matrix(cl, m)
    require("grid")
    grid.newpage(); vp <- viewport(w = .92, h = .92)
    grid.rect(gp=gpar(fill=bg))
    grid.text(cm, x = col(cm)/m, y = rev(row(cm))/m, rot = rot,
              vp=vp, gp=gpar(cex = cex, col = cm))
  }

showCols(bg="gray20",cl=colors()[1:60], rot=30, cex=0.9)
```

Los colores se pueden especificar mediante un código de color hexadecimal, como "#FF0000". ¿Qué color es este?

```r
plot(test, col="#05f936")
```

Vamos a usar otro conjunto de datos procedente del portal de datos abiertos de CYL: http://datosabiertos.jcyl.es/web/jcyl/set/es/mediciones/climatologia/1284225649821  

Además para leer caracteres acentuados y especificar el separador usado en el conjunto usamos los atributos ``fileEncoding`` y ``sep``

```r
clima <- read.csv(file = 'Clima_CYL_Diario.csv', sep=";", fileEncoding = "Latin1", check.names = F)
```

```r
str(clima)
```

Vamos a representar dos gráficas en la misma figura:

```r
temps <- clima[['Temperatura media (ºC)']]
precs <- clima[['Precipitación (mm)']]
```

```r
par(mfrow=c(1,2))    #establecer el área de dibujo en una matriz de 1 * 2
barplot(temps, main="Barplot")
pie(precs, main="Piechart", radius=1)
```
