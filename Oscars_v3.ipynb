```python
import pandas as pd
import altair as alt
import numpy as np

#prevent max_rows
alt.data_transformers.disable_max_rows()


oscars_all = pd.read_csv("data/the_oscar_award.csv")
icons = pd.read_csv("data/icons_category.csv", sep=";")


#name desambiguation found
oscars_all["name"] = oscars_all["name"].str.replace("Francis Coppola", "Francis Ford Coppola")

#split multiple awardee/nominee into several rows

oscars_all['name_single'] = oscars_all['name']
#multiple=oscars_all.loc[oscars_all['name_single'].str.contains(";", case=False)]
oscars_all["name_single"]=oscars_all["name_single"].str.split(pat="; ")
oscars_all=oscars_all.explode("name_single").reset_index(drop=True)

oscars_all["name_single"] = oscars_all["name_single"].str.replace("\n", " ")

oscars_all["name_single"] = oscars_all["name_single"].str.replace("Music and Lyric by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Music and Lyrics by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Music by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Lyric by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Lyrics by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Screeplay by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Story by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Interior Decoration: ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Set Decoration: ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Art Direction: ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Written by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Production Design: ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Photographic Effects by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Sound Effects by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Screenplay by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Adaptation Score by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Song Score by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Adaptation by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Based on a story by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Orchestral Score by ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Special Visual Effects ", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("musical director  ", "")



 






oscars_all["name_single"] = oscars_all["name_single"].str.replace("head of department", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("\(production company\)", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace(r'\(Score by ', "")

oscars_all["name_single"] = oscars_all["name_single"].str.replace(r'\)', "")



 





oscars_all["name_single"] = oscars_all["name_single"].str.replace(", Jr.", " Jr.")
oscars_all["name_single"] = oscars_all["name_single"].str.replace(", Inc.", " Inc.")


oscars_all["name_single"] = oscars_all["name_single"].str.replace(", Producers", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace(", Producer", "")

oscars_all["name_single"] = oscars_all["name_single"].str.replace(", Sound Director", "")

oscars_all["name_single"] = oscars_all["name_single"].str.replace("Pinewood Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Metro-Goldwyn-Mayer Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("20th Century-Fox Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Warner Bros. Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Samuel Goldwyn Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Columbia Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Paramount Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("RKO Radio Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Republic Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Universal Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Walt Disney Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Todd-AO Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Universal-International Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Hal Roach Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Samuel Goldwyn - United Artists Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Universal City Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("United Artists Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Shepperton Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Warner Bros.-First National Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Warner Bros.-Seven Arts Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("General Service Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Paramount Publix Studio Sound Department,", "")
oscars_all["name_single"] = oscars_all["name_single"].str.replace("Fox Studio Sound Department,", "")














oscars_all["name_single"]=oscars_all["name_single"].str.split(" and ")
oscars_all=oscars_all.explode("name_single").reset_index(drop=True)
oscars_all["name_single"]=oscars_all["name_single"].str.split(pat=", ")
oscars_all=oscars_all.explode("name_single").reset_index(drop=True)
oscars_all.drop_duplicates()







oscars_all['cat']='todefine'
oscars_all.loc[oscars_all['category'].str.contains('MUSIC', case=False), 'cat'] = 'MUSIC'
oscars_all.loc[oscars_all['category'].str.contains('CINEMATOGRAPHY', case=False), 'cat'] = 'CINEMATOGRAPHY'
oscars_all.loc[oscars_all['category'].str.contains('INTERNATIONAL', case=False), 'cat'] = 'FOREIGN'
oscars_all.loc[oscars_all['category'].str.contains('FOREIGN', case=False), 'cat'] = 'FOREIGN'
oscars_all.loc[oscars_all['category'].str.contains('COSTUME', case=False), 'cat'] = 'COSTUME'
oscars_all.loc[oscars_all['category'].str.contains('DOCUMENTARY', case=False), 'cat'] = 'DOCUMENTARY'
oscars_all.loc[oscars_all['category'].str.contains('DIRECTING', case=False), 'cat'] = 'DIRECTING'
oscars_all.loc[oscars_all['category'].str.contains('SHORT', case=False), 'cat'] = 'SHORT'
oscars_all.loc[oscars_all['category'].str.contains('BEST MOTION PICTURE', case=False), 'cat'] = 'BEST MOTION PICTURE'
oscars_all.loc[oscars_all['category'].str.contains('OUTSTANDING PRODUCTION', case=False), 'cat'] = 'BEST MOTION PICTURE'
oscars_all.loc[oscars_all['category'].str.contains('OUTSTANDING MOTION', case=False), 'cat'] = 'BEST MOTION PICTURE'
oscars_all.loc[oscars_all['category'].str.contains('OUTSTANDING PICTURE', case=False), 'cat'] = 'BEST MOTION PICTURE'
oscars_all.loc[oscars_all['category'].str.contains('UNIQUE', case=False), 'cat'] = 'BEST MOTION PICTURE'
oscars_all.loc[oscars_all['category'].str.contains('BEST PICTURE', case=False), 'cat'] = 'BEST MOTION PICTURE'
oscars_all.loc[oscars_all['category']== "ACTRESS", 'cat'] = 'ACTRESS'
oscars_all.loc[oscars_all['category']== "ACTRESS IN A LEADING ROLE", 'cat'] = 'ACTRESS IN A LEADING ROLE'
oscars_all.loc[oscars_all['category']== "ACTRESS IN A SUPPORTING ROLE", 'cat'] = 'ACTRESS IN A SUPPORTING ROLE'
oscars_all.loc[oscars_all['category']== "ACTOR", 'cat'] = 'ACTOR'
oscars_all.loc[oscars_all['category']== "ACTOR IN A LEADING ROLE", 'cat'] = 'ACTOR IN A LEADING ROLE'
oscars_all.loc[oscars_all['category']== "ACTOR IN A SUPPORTING ROLE", 'cat'] = 'ACTOR IN A SUPPORTING ROLE'
oscars_all.loc[oscars_all['category'].str.contains('SOUND', case=False),'cat'] = 'SOUND'
oscars_all.loc[oscars_all['category'].str.contains('SPECIAL', case=False),'cat'] = 'SPECIAL EFFECTS'
oscars_all.loc[oscars_all['category'].str.contains('WRITING', case=False),'cat'] = 'WRITING'
oscars_all.loc[oscars_all['category'].str.contains('MAKEUP', case=False),'cat'] = 'MAKEUP'
oscars_all.loc[oscars_all['category'].str.contains('ART DIRECTION', case=False),'cat'] = 'ART DIRECTION'
oscars_all.loc[oscars_all['category'].str.contains('HONORARY AWARD', case=False),'cat'] = 'HONORARY AWARD'
oscars_all.loc[oscars_all['category'].str.contains('EDITING', case=False),'cat'] = 'EDITING'
oscars_all.loc[oscars_all['category'].str.contains('VISUAL EFFECTS', case=False),'cat'] = 'VISUAL EFFECTS'
oscars_all.loc[oscars_all['category'].str.contains('ANIMATED FEATURE FILM', case=False),'cat'] = 'ANIMATED FEATURE FILM'
oscars_all.loc[oscars_all['category'].str.contains('JEAN HERSHOLT HUMANITARIAN AWARD', case=False),'cat'] = 'JEAN HERSHOLT HUMANITARIAN AWARD'
oscars_all.loc[oscars_all['category'].str.contains('ASSISTANT DIRECTOR', case=False),'cat'] = 'ASSISTANT DIRECTOR'
oscars_all.loc[oscars_all['category'].str.contains('PRODUCTION DESIGN', case=False),'cat'] = 'PRODUCTION DESIGN'
oscars_all.loc[oscars_all['category'].str.contains('DANCE DIRECTION', case=False),'cat'] = 'DANCE DIRECTION'
oscars_all.loc[oscars_all['category'].str.contains('ENGINEERING EFFECTS', case=False),'cat'] = 'ENGINEERING EFFECTS'
oscars_all.loc[oscars_all['category'].str.contains('THALBERG', case=False),'cat'] = 'IRVING G. THALBERG'

oscars_all = oscars_all.set_index('category')
icons = icons.set_index('category')
oscars_all['url']=icons['url']
oscars_all.reset_index(inplace=True)
oscars_all['url']=oscars_all['url'].replace(np.NaN,"holder")


counts=oscars_all['name_single'].value_counts().rename_axis('name_single').to_frame('total')
counts=counts.reset_index()

oscars_all = oscars_all.set_index('name_single')
counts = counts.set_index('name_single')
oscars_all['total']=counts['total']
oscars_all.reset_index(inplace=True)

oscars_all['film']=oscars_all['film'].replace(np.NaN,"N/A")


oscars_all=oscars_all.sort_values(by=['total'],  ascending=False)










########## FILTERING
threshold=15
sel =  oscars_all['total'].astype(int)>threshold

#sel=(oscars_all['total'].astype(int)>4) & (oscars_all['cat']=='DIRECTING')
#################

res=oscars_all.loc[sel].dropna()

res['name_single'] = res['name_single'].str.strip()


second_set=pd.DataFrame(res.groupby('name_single').name, columns=['name_single', 'bla'])
first_years=res.groupby('name_single').year_ceremony.min().tolist()
last_years=res.groupby('name_single').year_ceremony.max().tolist()
total=res.groupby('name_single').total.max().tolist()


second_set['first_year']=first_years
second_set['last_year']=last_years
second_set['total']=total

oscars=pd.DataFrame(res.loc[res['winner']==True].groupby('name_single').winner.count())
second_set = second_set.set_index('name_single')
#oscars = oscars.set_index('name')
second_set['oscars']=oscars['winner']
second_set.reset_index(inplace=True)
second_set['oscars']=second_set['oscars'].replace(np.NaN,0.0)


res = res.set_index('name_single')
second_set = second_set.set_index('name_single')
res['oscars']=second_set['oscars']
res.reset_index(inplace=True)
second_set.reset_index(inplace=True)


sortedcat=sorted(oscars_all['cat'].unique())
range_cat=['#3366c3','#5a7ec2','#7d95c4','#f94d4d','#fd7575','#f79e9e',
           '#578449','#744984','#ec7805','#05ece8','#33ec05','#05b2ec',
           '#02516c','#6c2902','#68026c','#026c1c','#607966','#c14d09',
           '#ec6dd7','#6decb7','#0a736c','#730a50','#6c730a','#9c7c46',
           '#02516c','#6c2902','#68026c','#026c1c','#607966','#c14d09']


sortedres=sorted(res['cat'].unique())
indices = [sortedcat.index(x) for x in sortedres]
rangeres= [range_cat[x] for x in indices]



#theme creation
# define the theme by returning the dictionary of configurations
def oscar_theme():
    background= 'snow',
    font="Dosis",
    axisColor= '#cccccc',
    labelFont="Dosis",
    legendFont='Dosis',
    titleFont='Amity Jack'#'Noto Sans',
    gridColor="#f3cbc8"
    
    return {
        'background': background,
        'margin': 20,
        'config': {
            "title": {
                "fontSize": 18,
                "font": titleFont,
                "anchor": "start", # equivalent of left-aligned.
                "color": "#6b788c",
                "titleAchor": "middle",
                "dx": 115,
                "dy": -20,
                "titleAlign": "center"
            },
            "axisX": {
                "domain": True,
                "domainColor": axisColor,
                "domainWidth": 1,
                "grid": True,
                "labelFont": labelFont,
                "labelFontSize": 12,
                "labelAngle": 0, 
                "tickColor": axisColor,
                "tickSize": 5, # default, including it just to show you can change it
                "titleFont": font,
                "titleFontSize": 12,
                "titlePadding": 10, # guessing, not specified in styleguide
                "title": "X Axis Title (units)"
            },
            "axisY": {
                "domain": False,
                "grid": True,
                "gridColor": gridColor,
                "gridWidth": 1,
                "labelFont": labelFont,
                "labelFontSize": 12,
                "labelAngle": 0, 
                "ticks": False, # even if you don't have a "domain" you need to turn these off.
                "titleFont": font,
                "titleFontSize": 12,
                "titlePadding": 10, # guessing, not specified in styleguide
                "title": "Y Axis Title (units)", 
                # titles are by default vertical left of axis so we need to hack this 
                "titleAngle": 0, # horizontal
                "titleY": -10, # move it up
                "titleX": 18, # move it to the right so it aligns with the labels 
            },
            "legend": {
                "labelFontSize": 8,
                "labelFont": legendFont,
                "symbolSize": 100,
                "titleFontSize": 12,
                "titlePadding": 10,
                "titleFont": legendFont,
                "orient": "top-right",
                "offset": 5
              }
        }
    }
# register the custom theme under a chosen name
alt.themes.register('oscar_theme', oscar_theme)
# enable the newly registered theme
alt.themes.enable('oscar_theme')


numOfRows = res['name_single'].unique().shape[0]
row_height = 20
total_height= numOfRows*row_height
total_width=880

selection = alt.selection_multi(fields=['cat'], bind='legend')
brush = alt.selection(type='interval', encodings=['x'])


res=res.sort_values(by=['oscars','total','year_ceremony'],  ascending=[False, False, False])
second_set=second_set.sort_values(by=['oscars','total','first_year'],  ascending=[False, False, False])

manual_ordered = second_set.reset_index()
manual_ordered = manual_ordered.sort_values(['oscars','index'], ascending=[False, True])
manual_ordered = res.reset_index()
manual_ordered = manual_ordered.sort_values(['oscars','index'], ascending=[False, True])

the_title="And (with " + str(numOfRows) + " awardees with "+ str(threshold) +" nominations at least) the oscar goes to..."


base=alt.Chart(res).encode(
  #sorting does not work in altair, but works in the exported Vega-lite
  alt.Y('name_single:N', sort=alt.SortField(field="oscars", order="ascending")),
  alt.X('Year:T'),
  tooltip=(alt.value(None))
).transform_calculate(
    Year='datetime(datum.year_ceremony, 1)'
).properties(
    width=total_width,
    height=total_height,
    title=the_title
)


es=base.mark_boxplot().encode(
  color=alt.value('#b4c8e7'),
  size=alt.value(10), 
  tooltip=alt.value(None),
  opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_selection(
    selection
)

po=base.mark_circle().encode(
  alt.X('Year:T'),
  
  color=alt.condition(brush, alt.Color('cat:N', 
                  scale=alt.Scale(range=rangeres,
                                  domain=sortedres
                                 )
                 ), alt.value('lightgray')),
  opacity=alt.condition(selection, alt.value(1), alt.value(0.2)),
  tooltip=alt.Tooltip(['year_film', 'year_ceremony', 'category', 'winner','name','film','cat','total','name_single', 'oscars:Q']),
  size=alt.value(80)
).transform_calculate(
).add_selection(
    selection
)

base_label=alt.Chart(second_set).encode(
  alt.Y('name_single:N', axis=alt.Axis(labels=True),  sort=alt.SortField(field="oscars", order="ascending")),
  alt.X(axis=alt.Axis(labels=False), scale=alt.Scale(domain=[1926, 2021])),
  color=alt.value('#6b788c'),
  size=alt.value(8)
).transform_calculate(
    firstYear='datetime(datum.first_year, 1)',
    lastYear='datetime(datum.last_year, 1)'
)

label_name=base_label.mark_text(align='left', dx=-5, dy=-8,font='Amity Jack').encode(
  alt.X('firstYear:T'),
  text='name_single:N',
  tooltip=alt.Tooltip(['name_single', 'first_year','total','oscars','last_year']),
  opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
)

label_total=base_label.mark_text(align='left', dx=5, dy=-5, font='Amity Jack').encode(
  alt.X('lastYear:T'),
  text=alt.Text('oscars:Q'),
  size=alt.value(16), 
  opacity=alt.condition(selection, alt.value(1), alt.value(0.2)),
  color=alt.condition(brush, alt.value('#b0bbca'), alt.value('lightgray'))

)

label_oscar=base_label.mark_text(align='left', dx=4, dy=+5, font='Amity Jack').encode(
  alt.X('lastYear:T'),
  text=alt.Text('total:Q'),
    size=alt.value(10),
  color=alt.condition(brush, alt.value('#6b788c'), alt.value('lightgray')),
                 
  opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
)

oscar=base.mark_point(filled=True).encode(
  size=alt.value(290.0),
  color=alt.condition('datum.winner',
                      alt.value('#ffe100'),
                      alt.value(None)),
  opacity=alt.condition(selection, alt.value(0.5), alt.value(0.1))
)


timeline=es + label_name + label_total +label_oscar  +oscar +po


# Define the base chart, with the common parts of the
# background and highlights
base2 = alt.Chart().mark_bar().encode(
    x=alt.X(alt.repeat('column'), type='quantitative', bin=alt.Bin(maxbins=100), axis=alt.Axis(format=",.0f", labels=True)),
    y='count()'
).properties(
    width=total_width/2.1,
    height=130
)

# gray background with selection
background = base2.encode(
    color=alt.value('#cb274b')
).add_selection(brush)

# blue highlights on the transformed data
highlight = base2.transform_filter(brush)

# layer the two charts & repeat
stats=alt.layer(
    background,
    highlight,
    data=second_set
).transform_calculate(

).repeat(column=["total", "oscars"])

alt.vconcat(stats, timeline)
```
